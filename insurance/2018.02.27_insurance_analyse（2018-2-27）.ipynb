#### 每年存入固定金额保费（Money），共存N年（HowManyYears），保障M年（HowManyYears+AgainYears）
* DepositCumulationMoneyEveryYears(Y5,Y3,Y2,Y1,Money,HowManyYears)
* DepositCumulationMoneyAgainYears(Y5,Y3,Y2,Y1,Money,MoneyAgainYears,HowManyYears,AgainYears)
