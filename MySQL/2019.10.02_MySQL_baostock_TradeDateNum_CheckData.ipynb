```python
import pandas as pd
import pymysql
import DB_Select_Local as DBsl
import time
```

```python
def DB_Select(stock):
    
    #### 连接database ####
    # conn = pymysql.connect(host=“你的数据库地址”, user=“用户名”,password=“密码”,database=“数据库名”,charset=“utf8”)
    conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='123456' ,charset='utf8')
    # 使用cursor()方法获取操作游标，cursor()得到一个可以执行SQL语句的光标对象
    cursor = conn.cursor()    
    
    #### 选择数据库：stock ####
    Select_DataBase = "use stock;"
    cursor.execute(Select_DataBase)
    table_name = stock
    sql_select = """select * from %s"""% table_name
    df = pd.read_sql(sql_select, con=conn)
    # 关闭光标对象
    cursor.close()
    # 关闭数据库连接
    conn.close()
    
    return df
```

```python
hs300 = hs300[['id','bs_date','bs_code','stock_name','weight','ipoDate','ts_TradeCount']]
hs300
```

```python
hs300['bs_TradeCount_1'] = ''
hs300['bs_TradeCount_2'] = ''
hs300['bs_TradeCount_3'] = ''
```

```python
for i in range(0,len(hs300)):
    stock_data = DBsl.DB_Select(hs300['bs_code'][i])
    
    stock_data_1 = stock_data.loc[(stock_data['adjustflag'] == '1')&(stock_data['frequency'] == 'd')&(stock_data['tradestatus'] == '1')]
    hs300.loc[i,'bs_TradeCount_1'] = len(stock_data_1)
    print(i,hs300['bs_code'][i],'后复权',len(stock_data_1))
    
    stock_data_2 = stock_data.loc[(stock_data['adjustflag'] == '2')&(stock_data['frequency'] == 'd')&(stock_data['tradestatus'] == '1')]
    hs300.loc[i,'bs_TradeCount_2'] = len(stock_data_2)
    print(i,hs300['bs_code'][i],'前复权',len(stock_data_2))
    
    stock_data_3 = stock_data.loc[(stock_data['adjustflag'] == '3')&(stock_data['frequency'] == 'd')&(stock_data['tradestatus'] == '1')]
    hs300.loc[i,'bs_TradeCount_3'] = len(stock_data_3)
    print(i,hs300['bs_code'][i],'不复权',len(stock_data_3),'\n')
```

```python
hs300['ts_TradeCount'] = hs300['ts_TradeCount'].astype(float)
hs300['check_data'] = hs300['ts_TradeCount']-(hs300['bs_TradeCount_1']+hs300['bs_TradeCount_2']+hs300['bs_TradeCount_3'])/3
```
