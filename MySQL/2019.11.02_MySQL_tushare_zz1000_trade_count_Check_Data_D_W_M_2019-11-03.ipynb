```python
def format_seconds(s):
    s = int(s)
    hh = s // 3600
    s = s % 3600
    mm = s // 60
    s = s % 60
    return str(hh)+':'+str(mm)+':'+str(s)
```

```python
def DB_Select(code):
    
    #### 连接database ####
    # conn = pymysql.connect(host=“你的数据库地址”, user=“用户名”,password=“密码”,database=“数据库名”,charset=“utf8”)
    conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='root' ,charset='utf8')
    # 使用cursor()方法获取操作游标，cursor()得到一个可以执行SQL语句的光标对象
    cursor = conn.cursor()
    
    #### 选择数据库：stock ####
    Select_DataBase = "use stock;"
    cursor.execute(Select_DataBase)
    table_name = code.replace(".", '')
    
    sql_select = """select * from %s"""% table_name
    
    df = pd.read_sql(sql_select, con=conn)
    
    return df
```

```python
def database_stock(stock):
    #### 连接database ####
    # conn = pymysql.connect(host=“你的数据库地址”, user=“用户名”,password=“密码”,database=“数据库名”,charset=“utf8”)
    conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='root' ,charset='utf8')
    # 使用cursor()方法获取操作游标，cursor()得到一个可以执行SQL语句的光标对象
    cursor = conn.cursor()
    #### 选择数据库：stock ####
    Select_DataBase = "use stock;"
    cursor.execute(Select_DataBase)
    
#     # conn = pymysql.connect(host=“你的数据库地址”, user=“用户名”,password=“密码”,database=“数据库名”,charset=“utf8”)
#     conn = pymysql.connect(host='122.112.201.53', port=3306, user='society_dev', passwd='FeidouDev@2019' ,charset='utf8')
#     # 使用cursor()方法获取操作游标，cursor()得到一个可以执行SQL语句的光标对象
#     cursor = conn.cursor()
#     #### 选择数据库：stock ####
#     Select_DataBase = "use society_dev;"
#     cursor.execute(Select_DataBase)
    
    table_name = 'zz1000'
    
    print('正在创建数据表：'+table_name)
    # 定义要执行的SQL语句，预处理语句创建表
    try:
        sql_create = """create table %s"""% table_name +""" (
                        id int auto_increment,
                        index_code VARCHAR(20),
                        con_code VARCHAR(20),
                        trade_date VARCHAR(20),
                        weight VARCHAR(10),
                        ts_TradeCount_D VARCHAR(20),
                        ts_TradeCount_W VARCHAR(20),
                        ts_TradeCount_M VARCHAR(20),
                        Count_D_hfq VARCHAR(20),
                        Count_D_qfq VARCHAR(20),
                        Count_D_bfq VARCHAR(20),
                        Count_W_hfq VARCHAR(20),
                        Count_W_qfq VARCHAR(20),
                        Count_W_bfq VARCHAR(20),
                        Count_M_hfq VARCHAR(20),
                        Count_M_qfq VARCHAR(20),
                        Count_M_bfq VARCHAR(20),
                        check_data_D VARCHAR(20),
                        check_data_W VARCHAR(20),
                        check_data_M VARCHAR(20),
                        PRIMARY KEY (id),
                        UNIQUE KEY (index_code,con_code,trade_date,code)
                        )"""
        # 执行SQL语句
        cursor.execute(sql_create)
    except:
        print(table_name+'数据表已存在！')
        
    length = len(stock)
    sql_insert = ''
    print('数据量：', length)
    for i in range(0, length):
        record = tuple(stock.loc[i])
        if sql_insert == '':
            #插入数据语句
            sql_insert = """REPLACE INTO  %s"""% table_name +""" (index_code,con_code,trade_date,weight,
                            ts_TradeCount_D,ts_TradeCount_W,ts_TradeCount_M,Count_D_hfq,Count_D_qfq,Count_D_bfq,
                            Count_W_hfq,Count_W_qfq,Count_W_bfq,Count_M_hfq,Count_M_qfq,Count_M_bfq,check_data_D,check_data_W,check_data_M) 
                            VALUES ('%s','%s','%s','%s','%s','%s','%s','%s','%s','%s','%s','%s',
                            '%s','%s','%s','%s','%s','%s','%s')""" % record
        else:
            sql_insert = sql_insert + """,('%s','%s','%s','%s','%s','%s','%s','%s','%s','%s','%s','%s',
                                         '%s','%s','%s','%s','%s','%s','%s')""" % record
        
    try:
        # 执行sql语句
        cursor.execute(sql_insert)
        # 提交到数据库执行
        conn.commit()
    except:
        # Rollback in case there is any error
        conn.rollback()
    
    print('已存入数据库！')    
    # 关闭光标对象
    cursor.close()
    # 关闭数据库连接
    conn.close()
```
