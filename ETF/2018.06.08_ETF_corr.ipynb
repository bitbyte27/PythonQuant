```python
import numpy as np
import pandas as pd
```

```python
def SZ_ETF(stock):
    stock = stock.set_index('date')
    stock.index = pd.to_datetime(stock.index)
    stock = stock.sort_index(axis=0, ascending=True)
    stadate = '2006-06-01'
    enddate = '2018-06-08'
    stock = stock[pd.to_datetime(stock.index) >= pd.to_datetime(stadate)]
    stock = stock[pd.to_datetime(stock.index) <= pd.to_datetime(enddate)]
    stock['yesterday_close'] = stock['close'].shift(1)
    stock.loc[0:1,'yesterday_close'] = stock['open']
    stock['range'] = (stock['close']-stock['yesterday_close'])/stock['yesterday_close']
    stock['amplitude'] = (stock['high']-stock['low'])/stock['yesterday_close']
    return stock
```

## 市场指数

```python
# 上证综指
stock_sz = pd.read_csv('data/999999.csv')
stock_sz = SZ_ETF(stock_sz)

# 沪深300指数
stock_sz300 = pd.read_csv('data/999300.csv')
stock_sz300 = SZ_ETF(stock_sz300)

# 上证50指数
stock_sz50 = pd.read_csv('data/999050.csv')
stock_sz50 = SZ_ETF(stock_sz50)

# 上证180指数
stock_sz180 = pd.read_csv('data/999180.csv')
stock_sz180 = SZ_ETF(stock_sz180)
```

## ETF基金

```python
# 180ETF（510180）
stock_180ETF = pd.read_csv('data/510180.csv')
stock_180ETF = SZ_ETF(stock_180ETF)

# 50ETF（51050）
stock_50ETF = pd.read_csv('data/510050.csv')
stock_50ETF = SZ_ETF(stock_50ETF)

# 300ETF(510300)
stock_300ETF = pd.read_csv('data/510300.csv')
stock_300ETF = SZ_ETF(stock_300ETF)
```
