```python
import pandas as pd
import numpy as np
import CCI_Profit_Functions
```

```python
# 设定显示最大的行数
pd.set_option("display.max_rows", 100)  
```

```python
ETF300 = pd.read_csv('ETF300.csv')
ETF300 = ETF300.set_index('datetime')
```

```python
# CCI策略
ETF300_CCI = ETF300.sort_index(axis=0, ascending=True)
```

```python
R_list = [8,10,14,18,27,46,69]
S_list = [8,10,14,18,27,46,69]
# R_list = [14]
# S_list = [14]
T_list = [0.015]
RST = pd.DataFrame()
```

```python
for R in R_list:
    for S in S_list:
        for T in T_list:
            CCI_para = [R, S, T]
            ETF300_CCI = CCI_Profit_Functions.Strategy_CCI(ETF300_CCI,CCI_para)
            FirstDF = ETF300_CCI.loc[:,['close','volume','CCI']]
            # CCI类策略中，即“buy_compare1”>“buy_compare2”；并将这些收盘价设置为买点（1）
            buy_compare1 = -100
            buy_compare2 = ETF300_CCI.loc[:,'CCI']
            # CCI类策略中，即“sell_compare1”<“sell_compare2”；并将这些收盘价设置为卖点（-1）。
            sell_compare1 = 100
            sell_compare2 = ETF300_CCI.loc[:,'CCI']
            # CCI_Profit_Functions:Strategy_Point_CCI
            SPD = CCI_Profit_Functions.Strategy_Point_CCI(FirstDF,buy_compare1,buy_compare2,sell_compare1,sell_compare2)
            Buy = 'close'
            Sell = 'close'
            cash = 10000
            brokerage = 2.5
            SPFo = CCI_Profit_Functions.Strategy_Profit_once(SPD,Buy,Sell,cash,brokerage)
            SPFs = CCI_Profit_Functions.Strategy_Profit_second(SPD,Buy,Sell,cash,brokerage)
            RST.loc[str(CCI_para), 'Profit_once'] = SPFo.iloc[-1]['Profit_real']
            RST.loc[str(CCI_para), 'Profit_sec'] = SPFs.iloc[-1]['Profit_real']
```
