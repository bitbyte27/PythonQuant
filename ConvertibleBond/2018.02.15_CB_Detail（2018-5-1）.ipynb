```python
import pandas as pd
import numpy as np
import tushare as ts
import matplotlib.pyplot as plt
```

```python
# 读取在CB_Transaction_Data_PLOT_F生成的持仓总表
CB_Transaction_Data_Cost = pd.read_csv('cbcsv/CB_Transaction_Data_Cost.csv', encoding='gbk')
CB_Transaction_Data_Cost
```

```python
# 可转债信息表，获取自东方财富网并整理，未来要尝试用爬虫方式获取数据
# http://quote.eastmoney.com/center/fullscreenlist.html#bp_8
CB_detail_data = pd.read_csv('cbcsv/CB_detail_data.csv', encoding='gbk')
CB_detail_data = CB_detail_data.fillna(0)
CB_detail_data.head()
```

```python
# 可转债回售情况表，获取自雪球网并整理，未来要尝试用爬虫方式获取数据
# https://www.jisilu.cn/data/cbnew/#put
CB_back = pd.read_csv('cbcsv/CB_back.csv', encoding='gbk')
CB_back = CB_back.fillna(0)
CB_back.head()
```

```python
# 可转债强赎情况表，获取自雪球网并整理，未来要尝试用爬虫方式获取数据
# https://www.jisilu.cn/data/cbnew/#redeem
CB_redemption = pd.read_csv('cbcsv/CB_redemption.csv', encoding='gbk')
CB_redemption.head()
```

```python
# 将以上四个表的可转债代码字段名称重新命名为cbcode
CB_Transaction_Data_Cost.rename(columns={u'证券代码':'cbcode'}, inplace = True)
CB_detail_data.rename(columns={u'转债代码':'cbcode'}, inplace = True)
CB_back.rename(columns={u'代码':'cbcode'}, inplace = True)
CB_redemption.rename(columns={u'代码':'cbcode'}, inplace = True)
# 设置一个cbcode的起息日与剩余年限表，作为与三个表的合同索引表
CB_Transaction_Data_Cost_Day = CB_Transaction_Data_Cost[['cbcode',u'起息日',u'剩余年限']]
```

```python
# 要注意定时手动更新东方财富网的可转债比价表，避免出现缺失数据的情况
CB_Cost_detail_data = pd.merge(CB_detail_data,CB_Transaction_Data_Cost_Day,how='right',on=['cbcode'])
CB_Cost_detail_data.tail()
```

```python
CB_back_detail_data = pd.merge(CB_Cost_detail_data,CB_back,how='left',on=['cbcode'])
```

```python
CB_redemption_detail_data = pd.merge(CB_Cost_detail_data,CB_redemption,how='left',on=['cbcode'])
```

```python
import matplotlib.dates as mdates
```

```python
CB_Cost_detail_data.tail()
```

### CB_Transaction_Data_Cost与CB_detail_data两表的日期检验

```python
CB_Cost_detail_data[u'检验A'] = (mdates.date2num(pd.to_datetime(CB_Cost_detail_data[u'起息日']).astype(np.object))
                               -mdates.date2num(pd.to_datetime(CB_Cost_detail_data[u'申购日期']).astype(np.object)))
CB_checkout_Day = CB_Cost_detail_data[[u'检验A']].T
CB_checkout_Day['sum'] = CB_checkout_Day.apply(lambda x: x.sum(),axis=1)
```

```python
CB_Cost_detail_data[u'检验A'].dtypes
```

```python
CB_checkout_Day['sum'].dtypes
```

```python
print '起息日（申购日）校验：',
print round(CB_checkout_Day['sum'].values,2)
# 检验值约等于0，则日期数据正确
```

### CB_back与CB_detail_data两表的转股价和回售价的检验

```python
CB_back_detail_data[u'检验B'] = ((CB_back_detail_data[u'转股价']+CB_back_detail_data[u'转股价格'])/2*CB_back_detail_data[u'回售触发比']
                                +CB_back_detail_data[u'回售触发价格'])/2-CB_back_detail_data[u'回售触发价']
```

```python
CB_back_detail_data[u'检验B'].dtypes
```

```python
CB_back_detail_data[[u'转债名称',u'转股价',u'转股价格',u'回售触发比',u'回售触发价',u'回售触发价格',u'检验B'
                     ,u'到期赎回价',u'回售价']].round(3).head()
```

```python
CB_checkout_Price = CB_back_detail_data[[u'检验B']].T
```

```python
CB_checkout_Price['sum'] = CB_checkout_Price.apply(lambda x: x.sum(),axis=1)
```

```python
print '转股价&回售触发价的校验：',
print round(CB_checkout_Price['sum'].values,3)
# 检验值约等于0，则日期数据正确
```

### CB_redemption与CB_detail_data两表的转股价与强赎价检验

```python
CB_redemption_detail_data[u'检验C'] = (((CB_redemption_detail_data[u'转股价']+CB_redemption_detail_data[u'转股价格'])/2
                                     *CB_redemption_detail_data[u'强赎触发比']+CB_redemption_detail_data[u'强赎触发价格'])/2
                                     -CB_redemption_detail_data[u'强赎触发价'])
```

```python
CB_checkout_redemption = CB_redemption_detail_data[[u'检验C']].T
```

```python
CB_checkout_redemption['sum'] = CB_checkout_redemption.apply(lambda x: x.sum(),axis=1)
```

```python
print '转股价&强赎触发价的校验：',
print round(CB_checkout_redemption['sum'].values,3)
# 检验值约等于0，则日期数据正确
```
