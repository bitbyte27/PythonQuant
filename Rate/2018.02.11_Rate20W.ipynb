```python
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
```

```python
Ningbo = 200000
Rate1 = 6.3
Rate2 = 8.4
Rate3 = 18.0
Rate4 = 13.2
Rate5 = 8.8525
```

```python
Payment1 = {'Date':[u'2018-01-20',u'2018-02-11',u'2018-03-11',u'2018-04-11',u'2018-04-20'],
           'Rate-NB':[0,770,980,1085,315]}
Payment2 = pd.DataFrame(Payment1)
```

```python
Payment2['Date2'] = Payment2['Date'].shift(1)
Payment2.loc[0:1,'Date2'] = u'2018-01-20'
Payment2['Date1'] = pd.to_datetime(Payment2['Date']).astype(np.object)
Payment2['Date2'] = pd.to_datetime(Payment2['Date2']).astype(np.object)
Payment2['Day'] = (mdates.date2num(Payment2['Date1'])-mdates.date2num(Payment2['Date2'])).astype(np.int)
Payment2['Rate-6.3'] = Ningbo*Rate1/100/360*Payment2['Day']
Payment2['Rate-8.4'] = Ningbo*Rate2/100/360*Payment2['Day']
Payment2['Rate-18.0'] = Ningbo*Rate3/100/360*Payment2['Day']
Payment2['Rate-13.2'] = Ningbo*Rate4/100/360*Payment2['Day']
Payment2['Profit'] = Payment2['Rate-13.2']-Payment2['Rate-6.3']
```

```python
Payment_A = Payment2.loc[1:7,['Date','Day','Rate-NB','Rate-6.3','Rate-8.4','Rate-18.0','Rate-13.2','Profit']]
Payment_A.round(2)
```
